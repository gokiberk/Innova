```python
from datetime import datetime, timedelta
from statsmodels.tsa.arima.model import ARIMA

# Read the CSV file
df = pd.read_csv("sundays.csv", parse_dates=['DATETIME'])

print(df)

# Extract the time from the DATETIME column
df['TIME'] = df['DATETIME'].dt.time

# Select the rows with the desired date from the reference day
reference_date = datetime(2020, 5, 31)
reference_rows = df[df['DATETIME'].dt.date == reference_date.date()]

# Define ARIMA model parameters
order = (1, 0, 0)  # (p, d, q)

# Iterate over each 15-minute interval
prediction_results = []
for i in range(0, 24 * 4):  # 24 hours * 4 (15-minute intervals)
    # Calculate the target datetime
    target_datetime = reference_date.replace(hour=i // 4, minute=(i % 4) * 15)

    # Find the corresponding row in the reference day data
    target_time = target_datetime.strftime("%H:%M")
    reference_row = reference_rows[reference_rows['TIME'].astype(str).str.contains(target_time)]

    if not reference_row.empty:
        # Extract the target data for ARIMA modeling
        target_data = reference_row['CPULOAD'].values

        # Fit ARIMA model
        model = ARIMA(target_data, order=order)
        model_fit = model.fit()

        # Make prediction for the current 15-minute interval
        prediction = model_fit.forecast()[0]

        # Store the prediction result
        prediction_results.append((target_datetime, prediction))
    else:
        # Handle missing data if the reference day does not have data for the current 15-minute interval
        prediction_results.append((target_datetime, None))

# Print the prediction results
for datetime, prediction in prediction_results:
    print("Datetime:", datetime, "Prediction:", prediction)
```
